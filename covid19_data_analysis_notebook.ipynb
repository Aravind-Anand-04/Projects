### Task 3.3: find maximum infection rate for all of the countries.

```python
countries = list(corona_ds_aggregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(corona_ds_aggregated.loc[c].diff().max())
corona_ds_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_ds_aggregated.head()
```

### Task 3.4: create a new dataframe with only needed column

```python
corona_data = pd.DataFrame(corona_ds_aggregated["max_infection_rate"])
```

```python
deaths_csv = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Covid19 Spread Data Analysis/covid19_deaths_dataset.csv")
deaths_csv.head()
```

```python
deaths_csv.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
deaths_csv.head()
```

```python
deaths_aggregated_csv = deaths_csv.groupby("Country/Region").sum()
deaths_aggregated_csv.drop(["Province/State"],axis=1,inplace=True)
```

```python
deaths_aggregated_csv.head()
```

```python
countries = list(deaths_aggregated_csv.index)
max_death_rates = []
for c in countries :
    max_death_rates.append(deaths_aggregated_csv.loc[c].diff().max())
deaths_aggregated_csv["max_death_rate"] = max_death_rates
```

```python
deaths_aggregated_csv.head()
```

```python
death_data = pd.DataFrame(deaths_aggregated_csv["max_death_rate"])
```

```python
death_data.head()
```

```python
data2 = death_data.join(happiness_report_csv,how="inner")
```

```python
data2.shape
```

```python
mask = data2["max_death_rate"] == 0.0
data2_cleaned = data2[~mask]
data2_cleaned.shape
```

```python
GDP = data2_cleaned["GDP per capita"]
deaths = data2_cleaned["max_death_rate"]
sns.regplot(x=GDP,y=np.log(deaths))
```

```python
social = data2_cleaned["Social support"]
deaths = data2_cleaned["max_death_rate"]
sns.regplot(x=social,y=np.log(deaths))
```
